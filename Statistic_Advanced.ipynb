### Question 1: What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?
#### Answer:
#### Type I error occurs when a true null hypothesis is rejected (false positive). Its probability is alpha.
#### Type II error occurs when a false null hypothesis is not rejected (false negative). Its probability is beta.
#### Type I errors lead to detecting an effect that does not exist.
#### Type II errors lead to missing an effect that actually exists.
#### Decision-making must balance both errors by selecting a proper significance level and sample size.


### Question 2: What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?
#### Answer:
#### The P-value is the probability of obtaining results as extreme as the observed data assuming the null hypothesis is true.
#### A small P-value (typically < 0.05) provides evidence against the null hypothesis.
#### A large P-value indicates insufficient evidence to reject the null hypothesis.
#### It does not represent the probability that the null hypothesis is true.


### Question 3: Explain the difference between a Z-test and a T-test, including when to use each.
#### Answer:
#### A Z-test is used when the population standard deviation is known and the sample size is large (n â‰¥ 30).
#### A T-test is used when the population standard deviation is unknown and the sample size is small (n < 30).
#### Z-test uses the standard normal distribution, while T-test uses the t-distribution.
#### T-test accounts for uncertainty in estimating the standard deviation from the sample.


### Question 4: What is a confidence interval, and how does the margin of error influence its width and interpretation?
#### Answer:
#### A confidence interval is a range of values likely to contain the true population parameter.
#### The margin of error determines the width of the interval.
#### A larger margin of error results in a wider interval, implying more uncertainty.
#### A smaller margin of error produces a narrower interval, implying higher precision.


### Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?
#### Answer:
#### ANOVA compares means of three or more groups to determine if at least one group mean differs significantly.
#### Assumptions include independence, normality of residuals, and equal variances across groups.
#### ANOVA extends hypothesis testing beyond two groups without increasing Type I error through multiple pairwise tests.


### Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.
#### Answer:


```python

import numpy as np
from scipy.stats import norm

data = np.array([52, 55, 50, 53, 54, 56, 51, 52])
mu_0 = 50
sigma = 3

sample_mean = np.mean(data)
n = len(data)

z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))
p_value = 1 - norm.cdf(z_stat)

print("Z-statistic:", z_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

### Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.
#### Answer:


```python
import numpy as np
import matplotlib.pyplot as plt

samples = 500
sample_size = 50
means = []

for _ in range(samples):
    data = np.random.exponential(scale=1.0, size=sample_size)
    means.append(np.mean(data))

plt.hist(means, bins=30)
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.title("Central Limit Theorem Demonstration (Exponential Distribution)")
plt.show()
```

### Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.
#### Answer:


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    n = len(data)

    z = norm.ppf((1 + confidence) / 2)
    margin = z * (std / np.sqrt(n))

    lower = mean - margin
    upper = mean + margin

    print("Mean:", mean)
    print("Confidence Interval:", (lower, upper))

    plt.errorbar(1, mean, yerr=margin, fmt='o')
    plt.xlim(0, 2)
    plt.title("Confidence Interval")
    plt.show()

data = np.random.randn(50)
confidence_interval(data)
```

### Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.
#### Answer:


```python
import numpy as np
from scipy.stats import chisquare

observed = np.array([30, 50, 20])
expected = np.array([33.3, 33.3, 33.3])

# Fix expected values so total matches observed total
expected = expected * (observed.sum() / expected.sum())

chi_stat, p_value = chisquare(observed, f_exp=expected)

print("Chi-square statistic:", chi_stat)
print("P-value:", p_value)
```

## 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_lis(lis):
    total=0
    for i in lis:
        if i%2==0:
            total+=i
    return total
lis =[2,4,6,78]
print(sum_lis(lis))
```

## 2. Create a Python function that accepts a string and returns the reverse of that string.

```python
def rev(st):
    l=len(st)
    s=''
    for i  in range(l-1,-1,-1):
        s+=st[i]
    return s

s="Awanit"
rev(s)
```

## 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

```python
def sq(nums):
    l=len(nums)
    for i in range (0,l):
        nums[i]=nums[i]**2
        
    return nums

num=[1,2,3,4,5,6]
sq(num)
```

## 4. Write a Python function that checks if a given number is prime or not from 1 to 200

```python
def prime(num):
    
    half=num//2
    if num==1 :
        return False
    for i in range(2,half+1):
        if num%i==0:
            return False
    return True
    
num=int(input("Enter the number: "))
print(prime(num))

#Work for all prime numbers
```

## 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number o terms.

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n          
        self.i = 0          
        self.a = 0          
        self.b = 1          

    def __iter__(self):
        return self         

    def __next__(self):
        if self.i >= self.n:
            raise StopIteration  
        if self.i == 0:
            self.i += 1
            return self.a
        elif self.i == 1:
            self.i += 1
            return self.b
        else:
            self.i += 1
            self.a, self.b = self.b, self.a + self.b  
            return self.b
        
        
fib_seq = Fibonacci(10)
for num in fib_seq:
    print(num, end=" ")
```

## 6. Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def power_of_two(n):
    for i in range (1,n+1):
        yield 2**i
        
n=int(input("Enter the maximum expoinent : "))
it=(power_of_two(n))
for i in range(0,n+1):
    print(next(it))
#atlast it exceed the n and give 'StopIteration'  error
```

## 7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(filename):
    
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()  

it=read_file_line_by_line("Example.txt")
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
```

## 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(st):
    for i in st:
        if i =='a' or i =='e' or i =='i' or i =='o' or i =='u':
            st=st.replace(i,"")
            
    return st
st="Hello I am Awanit Saha I am Currently Pursing  My Btech from Bennett University "
print(remove_vowels(st))
```
